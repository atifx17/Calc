```python
from tkinter import*
root=Tk()
root.geometry("500x600")
root.title("TO do list")

def add():
    x=e.get()
    if x:
        l.insert(END,x)
        e.delete(0,END)
        
   
    


def de():
    q=l.curselection()
    if q:
        l.delete(q)
    
def edit():
    q=l.curselection()
    if q:
        new=e.get()
        if new:
            l.delete(q)
            l.insert(q,new)
            e.delete(0,END)




e=Entry(root,font=("arial",20))
l=Listbox(root,selectmode=SINGLE,font=("arial",20))
b=Button(root,text="Add",font=("arial",15),command=add)
r=Button(root,text="Remove",font=("arial",15),command=de)
ed=Button(root,text="Edit",font=("arial",15),command=edit)





e.place(x=45,y=0,width=400)
l.place(x=45,y=150,width=400,height=400)
b.place(x=300,y=45)
r.place(x=100,y=45)
ed.place(x=220,y=45)














root.mainloop()
```
